```python
import numpy as np
import matplotlib.pyplot as plt

import pymongo 
import string
import re
import seaborn as sns
import string
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

import pyspark
from pyspark.sql import SparkSession, Window
from pyspark.sql.functions import col, rank, corr, when
```

```python
# Connecting to MongoDB 
client = pymongo.MongoClient("mongodb://localhost:27017")
db = client["ds4300"]
```

```python
cpi = db["cpis"]
cpi_results = cpi.find({}, {"year": 1, "annual_cpi":1, "_id":0})

years = []
cpis = []
for result in cpi_results:
    if (result.get("year") is not None) and (result.get("annual_cpi") is not None):
        years.append(result["year"])
        cpis.append(result["annual_cpi"])
```

```python
spark = SparkSession.builder.appName("corr_analysis").getOrCreate()
```

```python
df = spark.createDataFrame(zip(years, cpis), schema=['Year','Annual'])
df.show()
```

```python
# plotting annual CPI index from 2006 to 2022
plt.plot([row[0] for row in df.select("Year").collect()], [row[0] for row in df.select("Annual").collect()])
plt.xlabel("Year")
plt.ylabel("CPI")
plt.title("Annual CPI")
plt.show()
```

```python
tweets = db["tweets"]

# Defining and executing a query to obtain an aggregated list of tweets per year 
pipeline = [{'$project': {
            'tweet': 1,
            'year': {'$year': {'$dateFromString': {'dateString': '$date'}}}}},
    {'$group': {
            '_id': '$year',
            'tweets': {'$push': '$tweet'},
            'count': {'$sum': 1}}},
    {'$sort': {'_id': 1}}]

results = list(tweets.aggregate(pipeline))
```

```python
for result in results: 
    tweets = result['tweets'] 
    
    processed_tweets = []
    for tweet in tweets: 
        s = tweet.lower()
        s = s.translate(str.maketrans("", "", string.punctuation))
        s = re.sub(r'\d+', '', s)
        
        # Append the processed string to the output list
        processed_tweets.append(s)
    
     # Replace the original tweets with the processed tweets
    result['tweets'] = processed_tweets
    
print(results)
```

```python
analyzer = SentimentIntensityAnalyzer()

# defining an empty dictionary 
year_sentiment = {}

for result in results:
    # define year and list of tweets 
    year = result['_id']
    tweets = result['tweets'] 
    
    # use VADER to keep track of avg compound scores for a given year 
    sentiment_scores = [analyzer.polarity_scores(tweet)['compound'] for tweet in tweets] 
    year_sentiment[result['_id']] = (sum(sentiment_scores)/len(sentiment_scores)) * 100

# Create a DataFrame from the dictionary
yearly_df = spark.createDataFrame(list(year_sentiment.items()), schema=['Year', 'Score'])

# Print the resulting DataFrame
yearly_df.show()
```

```python
# joining the CPI and sentiment score dataframes
merge_df = df.join(yearly_df, on="Year")
merge_df.show()
```

```python
# correlation between sentiment score and annual CPI index over all 15 years
merge_df.corr("Annual","Score")

# we observe that the calculated correlation coefficient is weak and negative
```

```python
"""
Computes the rolling correlations over a shorter timeframe given by the window parameter. The window parameter represents
the timeframe in years. Displays the correlation table and plots the rolling correlation with the year.
"""
def rolling_correlation(window):
    corr_df = merge_df\
      .withColumn("window",rank().over(Window.orderBy("Year")))\
      .withColumn("Correlation", when(col("window") >= window, corr(col("Annual"), col("Score"))
                                     .over(Window.orderBy("Year")
                                           .rangeBetween(1-window, Window.currentRow))).otherwise(None))\
      .drop(col("window"))
    corr_df.show()
    years = [row[0] for row in corr_df.select("Year").collect()]
    corrs = [row[0] for row in corr_df.select("Correlation").collect()]
    plt.plot(years, corrs)
    plt.xlabel("Year")
    plt.ylabel("Correlation coefficient")
    plt.title(f"Correlation coefficients over {window} year rolling window")
    plt.axhline(y=0, color='black', linestyle='--')
    plt.show()
```

```python
rolling_correlation(3)
```

```python
rolling_correlation(4)
```
